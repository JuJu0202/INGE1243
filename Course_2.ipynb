# Position, velocity and acceleration

During last course we saw that some elementary dimensions at play in physics were mass, length and time. Amoung these dimensions, time is probably the most intriguing one. Its role in physics (e.g. in special or general relativity), the way it flows, its origin, etc. is (was and will be) at the origin of many theoretical considerations and is still an extensive source of research today. When we refer to **time** in the laws of physics we really mean **time intervals** (written $\Delta t = t_{2}-t_{1}$ [$T$]). To quantify these intervals, let us consider our first measurement instrument, the **stopwatch**. A stopwatch is used to measure the time interval between a starting event and a finishing event (e.g. a ball is dropped (start), the ball touches the floor (end)). 

In such experiment, we notice that the **position** of an object changes. The position of object is specified thanks to **coordinates** in a **coordinate system**. In mathematics, coordinates are a set of numbers that specify the position of a point in a coordinate system. To keep things simple, let's consider first the position of an object in a single dimension (1D) like a **train** on a railway. Let us look at the following figure:

<img title="a title" alt="Alt text" src="../Figures/Cours 2/train_experiment.jpg" width="900">

To be more precise, we consider the position of an abstract **point** that we can easily identify **on** the train. Arbitrarily, in order to analyse the **motion** of the train, we keep a **record** of the train center of mass's (depicted by the white point on the train) coordinate at different **random** instants (supposing the stopwatch is started before the train starts moving)(the center of mass's coordinates is 0). The **origin** (mathematical "0") of the **frame of reference** is the *central station*. The different coordinates are labeled on points located 1 meter from each other. Such record can have the following form (keeping only 2 digits after the decimal point): 

| $i$-th observation       | $t$ (s)         | $x$ (m) | 
|:-----:|:--------------:|:-----:|
|1| 0.00 |  0.00 |
|2| 1.12 |  0.00 |
|3| 2.07 |  0.00 |
|4| 3.46 |  1.10 |
|5| 4.23 |  2.58 |
|6| 5.11 |  5.02 |
|7| 6.57 |  10.86 |
|8| **7.62** |  **16.42** |

<div class="alert alert-block alert-info">
The position of an object is its physical location in space relative to an origin or other reference point in a defined coordinate system. A number line is a 1D (one dimensional) coordinate system. A real number that matches the location of a point along a number line is called a coordinate of the point. </div> 

Let's make few observations about this record: 

- The first one is that time **flows**. We have no way to prevent it from flowing. We cannot stop it, go back in time nor slow it down (as far as classics mechanics is concerned). **Position**'s coordinate on the other hand can stay unchanged (time cannot), it can change a lot or not between two successive observations, etc. 
- The second one is that there appear to be not clear **pattern** behind these numbers. The train's movement is not governed by a physical law. It's govern by the train driver pushing the accelerator as pleased. Eventhough it's an arbitrary motion, it is interesting to elaborate a framework to study the motion, namely **kinematics**.

<div class="alert alert-block alert-info">
Kinematics is a subfield of physics, developed in classical mechanics, that describes the motion of <b>points</b>, bodies (objects), and systems of bodies (groups of objects) without considering the forces that cause them to move (it is sometimes called the <b>"geometry of motion"</b>. </div>

- The third one is that if the reference point of the coordinates system changes, the absolute position changes. So we really have to keep track of the **displacements** between the $(i+1)$-th and the $i$-th measurement. It is written $(\Delta x)_{i}=x_{i+1}-x_{i}$: the amount of space traveled during the corresponding time interval $(\Delta t)_{i}=t_{i+1}-t_{i}$. Our previous table evolves as follow:

| $i$-th observation       | $\Delta t$ (s)         | $\Delta x$ (m) | 
|:-----:|:--------------:|:-----:|
|1| 0.00 |  0.00 |
|2| 1.12 |  0.00 |
|3| 0.95 |  0.00 |
|4| 1.39 |  1.10 |
|5| 0.77 |  1.48 |
|6| 0.88 |  2.44 |
|7| 1.46 |  5.84 |
|8| 1.05 |  5.56 |

It reads as follow: between the fourth ($(t_{4},x_{4})=(3.46,1.10)$) and the third measurement ($(t_{3},x_{3})=(2.07,0.00)$), a time lapse of 1.39 second (3.46-2.07) was required to cover 1.10 meter (1.10-0.00). We also have trivially that:

$$t_{8} = \sum_{i}^{8}(\Delta t)_{i}=\sum_{i=1}^{8}(t_{i+1}-t_{i})=7.62$$

and

$$x_{8} = \sum_{i}^{8}(\Delta x)_{i}=\sum_{i=1}^{8}(x_{i+1}-x_{i})=16.42$$

The first object that is studied in kinematics is called the **average velocity**. It is defined as follow: the average velocity between two measurements is given by the displacement $(\Delta x)_{i}$ divided by the corresponding time interval. It has the physical dimension of a length per time [$LT^{-1}$]. This quantity is defined **between** two measurements (all across the interval).

$$v_{av}=\frac{(\Delta x)_{i}}{(\Delta t)_{i}}$$

Applying this definition, we have:

| interval       | $v_{av}$ (m/s)  |
|:-----:|:--------------:|
|(1-2)| 0.00 |
|(2-3)| 0.00 |
|(3-4)| 0.79 |
|(4-5)| 1.92 |
|(5-6)| 2.77 | 
|(6-7)| 4.00 | 
|(7-8)| 5.29 | 

Using this definition, we can link two position measurements with the following **model relation**:

$$x_{i+1}=x_{i}+v_{av}\cdot\Delta t$$

We insist that this relation is simply a **model** to **interpolate** the observations but it gives **no extra informations** on the motion. At this point, let us put aside the **number representation** and go for a **graphical one**. Here is the position coordinate plotted at the different times we realized measurements.

As it is constructed, this average velocity during a time intereval cannot tell us how fast, or in what direction, the particle was moving at any given time during the interval. To describe the motion in full detail, we need to define the velocity at any specific instant of time. Doing so would require the realization of an **infinite sequence of measurements**. As the time intervals $\Delta t$ get smaller and smaller, the diplacements would get shorter and shorter **but** the **ratio** between the two would not necessarily vanish. On the contrary, **at the limit** where $\Delta t$ goes to zero, the average velocity becomes the **instantaneous velocity** (written $v$):

$$v=\lim_{\Delta t\to 0}v_{av}=\lim_{\Delta t\to 0}\frac{\Delta x}{\Delta t}$$

We can even go further in the notation since, to define the instanteneous velocity, the **complete set of positions** has to be known at **any instant of time**. In such case, the **position** can be defined as a **function of time** - $x(t)$ (it is a *surjective function* since any given position $x^{*}$ can be mapped from a time $t$ so that $x(t) = x^{*}$). 

The instantaneous velocity is thus expressed with the mathematical *total* **derivative with respect to time**. It is thus also a function of time.

$$v(t)=\frac{dx(t)}{dt}$$

For the rest of the course, when we use the term "velocity", we will aways mean the instantaneous velocity rather than the average velocity. 

This definition suggests a graphical interpretation. Let us consider an arbritrary motion whose position is specified by the blue line on the following figures. The instantaneous velocity at a given time is the **slope** of the position curve at this time. The following graphs show you in what sense the instantaneous velocity is the **limit** as time goes to zero of different average velocities computed with a fixed reference time. This whole reasoning is valid to define other quantities of interest in kinematics. Among these quantities, the **acceleration** of objects plays a central role. Just as velocity describes the rate of change of position with time, acceleration describes the **rate** of change of velocity with time. In the same reasoning as we did above, we can measure a finite set of instantaneous velocities of a moving object at random instants of time, and construct the **average** acceleration. 

$$a_{av}=\frac{(\Delta v)_{i}}{(\Delta t)_{i}}$$

Using this definition, we can link two velocity measurements with the following **model relation**:

$$v_{i+1}=v_{i}+a_{av}\cdot\Delta t$$

From the complete set of instanteneous velocities, we can also define an **instantaneous** acceleration.

$$a=\lim_{\Delta t\to 0}a_{av}=\lim_{\Delta t\to 0}\frac{\Delta v}{\Delta t}=\frac{dv(t)}{dt}$$

I advice you to try and do **by yourself** this step-by-step definition.

With acceleration, we can go one step further because, since velocity is the derivative with respect to time of the position, the acceleration is the second derivative of the position. We have:

$$a(t)=\frac{d}{dt}v(t)=\frac{d}{dt}\frac{dx(t)}{dt}=\frac{d^{2}x(t)}{dt^{2}}$$

Thanks to this relation, let us characterise a motion with constant acceleration.

## Motion with constant acceleration (in 1D)

The constant **instantaneous** acceleration is written $a$. It has the dimension [$LT^{-2}$] and in the SI sytem, the unit [$m/s^{2}$] (e.g. $a=13.65 m/s^{2}$). 

With a constant acceleration, the average and instantaneous acceleration are the same. Let us use the instantaneous velocity $v_{0}$ (called the **initial velocity**) at $t=0$ and the velocity $v(t)$ at an arbitrary instant $t$ to compute the average acceleration:

$$a_{av}=a=\frac{v(t)-v_{0}}{t-0}$$
 
This gives us the instantaneous velocity as a function of time:

$$v(t)=v_{0}+a\cdot t$$

To complete the description of the motion, let's find the position $x(t)$. To do so, as for initial velocity, the **initial position** - $x_{0}$ - is also required. With this quantity, at an arbitrary time, it is possible to compute the average velocity:

$$v_{av}=\frac{x(t)-x_{0}}{t-0}$$

It occurs that we can also get a second expression for the average velocity (only valid when the acceleration is constant). In the specific case where the acceleration changes at a constant rate, the average velocity during any time interval is simply the arithmetic average of the velocities at the beginning and at the end of the interval (no matter the limits of this interval). Let us consider the interval between zero and t. We get:

$$v_{av}=\frac{v_{0}+v(t)}{2}$$

All together, the three previous equations lead to the following kinematics:

$$x(t)=x_{0}+v_{0}t+\frac{at^{2}}{2}$$

This motion is ubiquitous in physics since, at the surface of the Earth, the acceleration caused by gravity is (approximativelly) **constant** for a *reasonable* height of the object. At different points on Earth's surface, the **free fall acceleration** ranges from 9.764 to 9.834 $m/s^{2}$. A conventional standard value is defined as 9.80665 $m/s^{2}$ (the reference value for this course). This acceleration is frequently refered to with the letter $g$.

This motion, eventhough simple to describe mathematically, can be really disturbing in term of "gut feeling". **In vacuum**, **free fall motion** is **independent of the mass of the object**. It is only a matter of acceleration ! Which is the same, no matter the mass, for different object at Earth's surface. A bowling ball and a feather, in a **vacuumed** environement touch the floor at the exact same time when launched from the same initial positions ($x_{0}$) with the same initial velocities ($v_{0}$). ## A peculiar kinematics

It is the year 2027, your business engineering skills conducted you in the entertainment industry. Amoung the financial opportunities of the year to come, is "Indiana Jones 6: the mysterious opal. You happen to read the following extract from the story board: "the mysterious opal, once activated, is an object whose **velocity equals its position coordinate**". Your intuition tells you that something will go wrong with the special effects required to animate such object but your **kinematics** knowlegdes are not that fresh in your memory.

In order to formalise your intuition, you start discussing with the scriptwriter. Here is the reasoning. 

Let's suppose the opal is located one meter (**initial position $x_{0}$**) from Mr. Jones and it starts moving in a straight line towards the larger position coordinates. What would be its position after 10 milliseconds ($\Delta t=0.01$ s) ? You start with a relativelly **small** time lapse in order for the **slope approximation** to be valid (draw a picture). You happen to find that the position 10 ms later is $x_{1}$ given by:

$$x_{1}=x_{0}+v_{0}\cdot \Delta t$$

But you know that "its velocity equals its position" ($v_{0}=x_{0}$) so you equivalently write:

$$x_{1}=x_{0}+x_{0}\cdot \Delta t=x_{0}\cdot(1+ \Delta t)$$

To get better insights on its kinematics, you keep going and find its position other 10 ms later:

$$x_{2}=x_{1}+v_{1}\cdot \Delta t=x_{1}\cdot(1+ \Delta t)$$

where you substitute $x_{1}$ by the quantity you found above:

$$x_{2}=x_{0}\cdot(1+ \Delta t)^{2}$$

With this logic, you are able to tell **approximativelly** the position after 1 second ($100\times10$ ms) :

$$x_{100}=x_{0}\cdot(1+ \Delta t)^{100}$$

It gives you the numerical value **2.7048**... ($1\times1.01^{100}$) meters. You **bless** the existence of your calculator since without it, you would have been forced to do the product $1.01\times1.01\times...\times1.01$ with a hundred factors (which is a bit **time demanding**). Not completelly satisfied, you want to be more accurate on the position after one second and reduce the time interval $\Delta t$ to 1 millisecond. You think that the **slope approximation** would be more valid that way (as pictured earlier on this course note). And you are right ! 

$x_{1000}=x_{0}\cdot(1+0.001)^{1000}$

It gives you the numerical value **2.7169**... meters (a single centimeter of change compared to your previous estimate).

After a little bit of reasoning, you find the position after 2 seconds. You recall a exponent algebraic property to write:

$$x_{2000}=x_{0}\cdot(1+0.001)^{2000}=x_{0}\cdot((1+0.001)^{1000})^{2}\approx 2.71^{2}$$

To compute the solution aterf 3, 4 or $t$ seconds (with $t$ designating the time since the opal is triggered), you always get:

$$x(t)\approx x_{0}\cdot 2.71^{t}$$

In order to lighten the notations, you call this constant that you tend to (**the smaller $\Delta t$ gets**): $e$, as a reference to your favorite mathematician Leonhard **E**uler. 

<div class="alert alert-block alert-info">
The number e, also known as Euler's number, is a mathematical constant approximately equal to 2.71828. The number e is of great importance in mathematics, physics, etc. alongside 0, 1, π, and i (complex number). Like the constant π, e is irrational (it cannot be represented as a ratio of integers) and transcendental (it is not a root of any non-zero polynomial with rational coefficients). </div>

In particular, as $\Delta t$ tends to zero, the **slope approximation** gets better and better. The answer **at the limit** is the following:

$$x(t)=x_{0}\cdot e^{t}$$

After 5 seconds, the opal is located approximatelly 150 meters away from Indiana... which is not that great for the cameraman and the dialogues.  

After 20 seconds, the opal has realized more than **12 times the world tour**. Supposing a mass of 1 kg, its kinetic energy would be of the order of $10^{5}$ Tera joules (as an indication, know that the maximal theoretical power/mass ratio of thermonuclear weapons is 25 Tera joules per kilogram). 

With these considerations, you change a little bit the project "Gantt chart" to include an extra working period for the scenarists. 

This little adventure also shows you how some **differential equations** (e.g. $v(t)=\frac{dx(t)}{dt}=x(t)$ with $x(t_{0})=v(t_{0})=x_{0}=1$) can be solved thanks to kinematics.
