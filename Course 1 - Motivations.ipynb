### The piano tuners in chicago 

One of the classic Fermi estimation problem is to estimate the number of piano tuners there are in the city of Chicago. Such problem presents so many unknown and the exact number seems (and perhaps is) unknowable. But actually, there is a way to get closer to it. As one may find, there are $10^{6}$ people in chicago. Supposing it is a rather young city, one could consider that *one out of a hunderd* ($10^{-2}$) people owns a piano. This means that there are roughly $10^{4}$ pianos in chicago that need to be tuned at least once a year. With really few researches on the complexity of the task, any good manager could estimate that tuning $10^{2}$ per year is a good rate for a tuner. Our back of the envelope computation tells us that there is *room* for $10^{2}$ piano tuners ($10^{6}\times10^{-2}\times10^{-2}$) to make a living. The fun and baffling fact is that at the time Fermi asked for this question, there was 82 piano tuners in the city of Chicago . Take a moment to appreciate the dialectic and the result.  

Fermi firmly believed that over-estimates and under-estimates would compensate and bring us at maximum one order of magnitude (o.o.m.) away from the true answer. 

The valuable economic toughts in this is that, with such computations, you can estimate weither or not the sector would be competive. As you can see there are not much more market shares for piano tunersleft (unless you revolutionize the piano tuning techniques). Also, as a piano tuner manager, I should know that there is *room* for tens or more workers but surely not a hundred or a thousand.   

Eventually let's recall that informations about o.o.m. are fairly accessible nowadays thanks to the internet, search engines  and social networks. Together with probability and statistics, this methodology opens a door on many answers.

# Physics : principles and laws

capturing how physics is at play and to domesticating the aforementioned motivations requires inevitably to grasp and understand the fundamental building blocks in physics. 

The elementary block in physics is **observations**. Primarily with your **bare senses** and then with **scientific instruments**. Once a sufficient amount of observations of a phenomenon have been collected, we can try to propose a **representation**. The different types of representations are only limited by mankind imagination but a common way to represent a phenomenon in physics is with the help of mathematical tools (numbers, algebra, calculus etc.) in the so-called **laws of physics**. These laws are then used to realise **predictions** that can be used in other disciplines such as engineering, chemistry, etc. The **validity** (truth criterion) of a physical law (or physical theory) is tested thanks to **experiments**. As soon as an observation contradicts the theory, it is no more considered as valid. In this sense, everybody can experiment physics (in both meanings of "to test" and "to challenge") and is encouraged to do so !

The above picture of physics (which is not exhaustive of course) is lacking a fundamental element : **principles**. The question is the following. where $n_1$ and $n_2$ are material parameters known as the **refractive index** of material 1 (air) and material 2 (water). From this point, you can keep developing technicalities such as the fact that the refractive index depends: on the color of the light beam (that is to say the color **wavelength** (a concept developed in chapter 8 of this course)), on temperature, etc. But the principal step is achieved: you have a **physical law** at your disposal. The optic fibers exploit **total internal reflection** to carry optic signals over long distances with little attenuation. These fibers are usually made with drawing glass (**silica**). When light is in the fiber, the material it is inside is silica (that has a refractive index of about 4 (for red light)). It is supposed to go into air (with a refractive index of 1). Let's write the law: 

$$4\times sin(\theta_{1})=1\times sin(\theta_{2})$$

As you know, the value of sine cannot get greater than 1 so the value of $sin(\theta_{1})$ cannot get greater than 1/4. A different physics is at play for value greater than 1/4: total reflexion and evanescent waves (waves that are not in the scope of this course).

The angle corresponding to 1/4 is called the **critical angle** ($\theta_{c}$) and as a value of 14.47° here. From Snell's law, we retrieve the following formula:

$$\theta_{c}=arcsin(\frac{n_{2}}{n_{1}})$$

# Dimensions, units and order of magnitude

Now that we discussed extensively about **conceptual physics**, it is time to start building a practical framework to **write down** physics. The starting point of such journey is the concept of **physical dimensions** of objects you can estimate with your bare senses: mass, length and time. Specifying that this dimensions are **physical** is important to avoid misconceptions mixing mathematical concepts developed also around the term "dimension". In some sense, in can be useful to make the distinction but from now, consider that word dimension stands for "physical dimension".   

<div class="alert alert-block alert-info">
The dimensions in this course are noted with <b>square bracket parentheses</b>. E.g. a object with dimension of mass is signified with [M], length with [L] and time with [T] </div> 

These dimensions distinguish physical objects from mathematical ones (pure numbers). When writing down physics, you come across **equations**. With a proper **dimensional analysis** you make sure that what you write has *at least* a physical sense. It is a very useful tool to check the correctness of formulas of various kinds and here is the principal piece of advice:

<div class="alert alert-block alert-info">
Always check the dimensional consistency of equations by making sure that the physical dimension of whatever is on the left hand side of an equation appears exactly on the right hand side as well </div> 

For example, you are covering a distance d (with dimension of length [$L$]) in a given time ([$T$]) and you are wondering how to relate your observations to your velocity. You may or not be moving at a constant speed but when you write down the equation expressing the velocity, whatever your write must *at least* have the dimension of length per time ([$LT^{-1}$]). It is call a **sanity check**.

Let's take another example, according to Newton's second law of motion, "force equals mass times acceleration". Acceleration is the **rate** at which velocity changes, so it has the physical dimensions [$LT^{-2}$]. A force has thus the dimension of mass times length times the square inverse of a time [$MLT^{-2}$] **and so is true for every force, no matter its nature (gravitational, electrical, frictional, etc.)**

With this in mind, you can now avoid a lot of mistakes. Consider the following. In an old book of physics, you happen to read that the gravitional force *equals* the product of the involved masses divided by the square of the distance between them". Mentaly, you can check that it has dimension of mass square divided by length square ([$M^{2}L^{-2}$]) which does not have the dimension of a force ! You can directly tell that this book is probably missleading. 

The power of dimensional analysis is not only to check for consistency. It can also help you to realize **back of the envelope computations** and capture insights about technologies.  

To this end, let's consider the valuable example of the **simple pendulum**. A simple pendulum is modeled as a rod of length $l$ [L] and a heavy bob of mass $m$ [M] and is free to oscillate about its point of suspension as presented on the following picture :

<img title="a title" alt="Alt text" src="../Figures/simple_pendulum.jpg" width="300">

The initial angular amplitude is $\theta_{0}$. This quantity is **adimensional** and is conventionaly expressed in **radians** in physics related problems.  

<div class="alert alert-block alert-info">
To measure an angle between two lines, you draw a circle centered at the intersection, no matter its radius (homothety). You measure the arc lenght defined by the lines (in blue on the following picture) and divide in by the radius of the circle (in red). The angle $\theta$ has the dimension of [L/L] which is noted <b>[-]</b> 
<img title="a title" alt="Alt text" src="../Figures/angle.jpg" width="300">
</div> 

The pendulum is supposed to be on **earth**. If we neglect friction (we do here), the oscillation continues forever. By conservation of energy, we take for a fact that the pendulum angular amplitude cannot get greater than the initial $\theta_{0}$. The mass of the rod is also neglected, if not, this is no more a so-called **simple** pendulum. 

The question is the following: What is the **time period** (noted $t^{*}$) of the oscillation ? The time period [T] is the time needed for the bob to come back at its initial position. The complete approach would be to solve Newton's equation of motion, show that it is a periodic motion and compute $t^{*}$. At this stage of the course, we are not "able" to realize this yet, but we could get closer to the final answer with only **dimensional analysis**. 

The first step could just be to list the quantity we think it could depend on. As a first guess, let's consider the following list: 

$m$,$l$,$g$ (known as the **standard gravity** at Earth's surface), $\theta_{0}$ (a first guess would be that the larger it gets, the longer it takes for the pendulum to comes back at the original position), c (light velocity in vacuum), h (Planck's constant), G (the gravitational constant)(indeed, maybe not only the Earth's gravity is relevant, maybe the sun, jupiter or the moon are playing a gravitational role), $k_B$ (Boltzman's constant) etc. 

Let's insist that it is a first **guess**. During the course, we managed to reduce this list with several physical arguments (e.g. Boltzman's constant $k_B$ is at play when temperature plays a role. It could indeed, due to thermal dilation, change the length of the rod and thus change a critical parameter of the problem. But it is **implicit** in such problem formulation that the temperature is fixed - otherwise, physical problem would take pages to be written down) to the following:

$m$,$l$,$g$,$\theta_{0}$

To **construct** an object that has the physical dimension of a time, we use the product of the listed quantities (with **integer** powers $a$,$b$ and $c$) *times* an **unknown** function $f$ of $\theta_{0}$. **Only dimensionless quantities can appear in function**.

<div class="alert alert-block alert-info">
A bogus argument is that real function can always be expended as a Taylor power series. For exemple, a function of the length $l$ would be non-sensical as we would add objects with the physical dimensions of length [$L^{1}$] to a surface [$L^{2}$] to a volume [$L^{3}$] etc.

$$f(l)\approx f(l_{0})+\frac{f'(l_{0})}{1!}(l-l_{0})^{1}+\frac{f''(l_{0})}{2!}(l-l_{0})^{2}+\frac{f'''(l_{0})}{3!}(l-l_{0})^{3}+ ...$$
    
The length could be at play in a function if another length ($l^{*}$) was *naturally* present in the problem so that we could *construct* a dimensionless quantity $\frac{l}{l^{*}}$.
</div> 

The time period can be written down:

$$t^{*}=m^{a}l^{b}g^{c}f(\theta_{0})$$

which in terms of dimensions reads:

$$[T]=[M]^{a}[L]^{b}[LT^{-2}]^{c}$$

recalling that $g$ has the dimension of an acceleration and the function $f(\theta_{0})$ is adimensional. 

The **only** (*only* is not justified - refer to the fundamental theorem of dimensional analysis if needed) way of constructing a time is then with the following constraints:

\begin{cases}
a=0 \\
b+c = 0 \\
-2c=1 \\
\end{cases}

The further we can go without Newton's equation is (with a=0,b=1/2 and c=-1/2):

$$t^{*}=f(\theta_{0})\sqrt{\frac{l}{g}}$$

What could this relation be used for if there remains such an indetermination regarding the $\theta_{0}$ dependency ?

Consider that a company plans on launching a simple pendulum to measure the absolute **time**. Using dimensional analysis, you can directly tell that this project will be a failure. Indeed, as you probably know, Earth is not a perfect sphere. For example, the radius at the pole is roughly 6350 km and at equator, 6372 km. With little research, you find out that Gravity depends on the **square**  the inverse distance between the **center of mass** of objects. Considering the rod length are equal (manufactured so) so as the initial angular amplitudes, the ratio between both time periods is given by:

$$\frac{t_{equator}}{t_{pole}}=\sqrt{\frac{g_{pole}}{g_{equator}}}=\sqrt{\frac{R^{2}_{equator}}{R^{2}_{pole}}}=\frac{R_{equator}}{R_{pole}}=1.003...[-]$$

Is this 3 after two 0s really a problem ? Yes ! Supposing the pendulum, with appropriately choosen physical parameters, gives you **one second** at the pole, it gives you 1.003... second at the equator. Which is a different of a bit more than **30 minutes after a week**: dramatic for many industries where time plays a central role (almost every company in fact). 

Now that we identifyied a convenient way to describe physical object -with respect to their dimensions - it is practical to use numbers to **compare** objects with the same dimension.  The numbers are expressed in a decimal basis and the **unit** for a given dimension is choosen **arbitrarily** and defined by a **standard** (or etalon) by the branch of physics called **metrology**. 

A famous intergovernmental organisation taking care of these precise definition is **The International Bureau of Weights and Measures (French: Bureau international des poids et mesures, BIPM)**, through which its 59 member-states act together on measurement standards in four areas: chemistry, ionising radiation, physical metrology, and coordinated universal time. 

Thanks to this international collaboration, a lot of scientists benefit from a common system of units - called the **International System of Units** (**SI sytem**). This is a widely used system but not the only one used (e.g. the imperial units) across the globe and it can be beneficial to use a different system of unit in specific scientist communities (e.g. studying atoms is conviniently done with the *atomic unit system*). The SI system defines **7** units related to **7** physical dimensions. Here is the list:
