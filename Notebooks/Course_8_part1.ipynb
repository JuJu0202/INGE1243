# Let's sum up 

Up to now we can divide what has been seen through this course into **2 sections** : the **tools** and their **application on motions**.

### The "tools" section

The tool category has been constructed with the following methodology. First we studied **Kinematics** thanks to coordinates, (unit) vectors, scalar product, instanteneous velocity and instanteneous acceleration. Second we went further, thanks to **Newton's laws of motion**, by studying motions that were not **arbitrary** : we studied **Dynamics**. From instantaneous kinematics quantities (e.g. velocity) and **physical properties** (e.g. the mass) we defined quantities that were ruled by Newton's laws (e.g. the linear momentum $\vec{p}=m\vec{v}$ whose time derivative is equal to the **net force** acting on a body). Eventually, we showed that the dynamical quantities could help us to define different forms of **Energy** (e.g. Potential, Kinetic, etc.). We showed how to benefit from the **Work-energy theorem** to get insights on daily life situations (e.g. skiding distance on a bike). 

This methodology was elaborated, in 2D, from both the $(x,y)$ coordinates (helping to define **linear** quantities e.g **linear momentum**) and the $(r,\theta)$ coordinates (helping to define **angular** quantities e.g. **angular momentum**).

### The work-energy theorem : the inclined plane - revisited

As one can see from the table presented in the begining  of this course, the **angular** equivalent of $\frac{mv^{2}}{2}$ is $\frac{I\omega^{2}}{2}$. Indeed, to change the angular velocity ($\omega=\frac{d\theta}{dt}$) of an object, **work** is required ! 

Understanding this analogy is left to you as an exercise. To help you, do not hesitate to consult chapter 10 of the reference book 'College Physics" (peculiarly section 10.4 : Rotational Kinetic Energy: Work and Energy Revisited).

Let us apply the work energy theorem to the problem of the inclined plane: 

$$\Delta K.E.=Work$$

This motion is really intersting as both the linear and angular velocity of the object change. Both inevitably change at a **rate** related to the mass $m$ and inertia $I$ of the rolling object. 

Here is a figure of the situation. A frame of reference is depicted on the picture but we will write a **scalar equation** thanks to the Work-Energy theorem. 

<img title="a title" alt="Alt text" src="../Figures/Cours 8/inclined.png" width="600">

From this picture, it is clear that the only force doing **work** (converting gravitational **potential** energy into **kinetic** energy) is the component $mgsin(\phi)$ of the gravitational force. Let us consider the distance $d$ as the distance covered by the object. The **work** expresses then : 

$$Work=mgsin(\phi)\times d$$

Let us now figure out **how** the Kinetic energy changes assuming we start with $v_{0}=0$ and $\omega_{0}=0$. We have :

$$\Delta K.E.=\frac{mv^{2}}{2}+\frac{I\omega^{2}}{2}$$

where $v$ is the **linear** velocity after the object covered a distance $d$ and $\omega$, its **angular** velocity after the same instant. The object has a mass $m$ and an inertia $I$.

The **scalar** equation based on the work-energy theorem is : 

$$\frac{mv^{2}}{2}+\frac{I\omega^{2}}{2}=mgsin(\phi)\times d$$

Now, from course 5, we know that the inertia is expressed as : $bmR^{2}$ where $b$ is an adimensional constant and $R$ the radius of the object (in this course, we only consider the inertia of "rolling" object : disk, sphere, hoop etc.).

From the same course, we know that, to rotate without slipping, the **linear** velocity and the **angular** one have to be related to each other with the simple expression:

$$v=R\omega$$

We can now re-write the equation in the following way :

$$\frac{mv^{2}}{2}+\frac{bmR^{2}}{2}\left(\frac{v}{R}\right)^{2}=mgsin(\phi)\times d$$

which is :

$$\left(\frac{1+b}{2}\right)mv^{2}=mgsin(\phi)d$$

which is equivalent to :

$$v^{2}(t)=\left(\frac{2gsin(\phi)}{1+b}\right)d(t)$$

Now comes the little trick : use the time derivative on both sides of this equation.

$$\frac{d}{dt}\left(v^{2}(t)\right)=\frac{d}{dt}\left(\left(\frac{2gsin(\phi)}{1+b}\right)d(t)\right)$$

which gives : 

$$2v(t)\frac{dv(t)}{dt}=\left(\frac{2gsin(\phi)}{1+b}\right)\frac{d\;(d(t))}{dt}$$

We can make **two** observations at this stage :

1. $\frac{dv(t)}{dt}$ is the **linear** acceleration $a$
2. \frac{d\;(d(t))}{dt} is the **linear** velocity $v$

By dividing by $v$ both sides of the equation, we get : 

$$a=\frac{gsin(\phi)}{1+b}$$

Which is exactly the same result as the one presented in course 5 ! ($\phi$ is considered positive in this case - it is just a convention)

We got this result with a **single** scalar equation (how powerful !). 

This approach will be tested during the exercise session 6 on the **simple pendulum** (going from **dimensional analysis** to **vector equations** and eventually applying the **work energy theorem**).
