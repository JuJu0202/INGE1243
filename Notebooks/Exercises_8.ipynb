# Physics can be used to play serious games

Keywords : metabolism, fluorine-18, PET scan, cyclotron, Lorentz force, Fermat spiral. 

The understanding of this video is the goal of this exercise session : https://www.youtube.com/watch?v=yrTy03O0gWw.

Cancer is a group of diseases involving abnormal cell growth with the potential to invade or spread to other parts of the body. A series of growth disorders can occur at the cellular level and these consequently underpin much of the subsequent course in cancer, in which a group of cells display uncontrolled growth and division beyond the normal limits, invasion (intrusion on and destruction of adjacent tissues), and sometimes metastasis (spread to other locations in the body via lymph or blood). Several key determinants of cell growth, like the regulation of cellular metabolism, are commonly disrupted in tumors. Cellular metabolism is the set of life-sustaining chemical reactions in organisms. The three main functions of metabolism are: the conversion of the energy available in food to energy available to run cellular processes; the conversion of food to building blocks for proteins, lipids, nucleic acids, and some carbohydrates; and the elimination of metabolic wastes.

In the 1920s, Otto Warburg observed that tumor cells consumed a large amount of **glucose**, much more than normal cells, and converted most of it to lactic acid. This phenomenon, now known as the “Warburg effect,” is the foundation of one of the earliest general concepts of cancer: that a fundamental disturbance of cellular metabolic activity is at the root of tumor formation and growth.

The aim of this exercise session is not to debate about the different treatments and different types of cancer. Nor to discuss the overall complexity of this problem. The goal is to try to answer the two following questions :

1. Are we able to **discriminate between** a well-functioning and a disfunctionning cell ?
2. Can we **locate** a disfunctioning cells in a living being ?

The strategy goes as follow: 

1. Tumor cells consume a large amount of **glucose**. 
2. Let us **attach** "something" to glucose that accumulates **where** glucose is highly metabolized (within the tumor cells). 
3. Let that "something" **reacts** and let us **trace back** its reaction **location**. 

This strategy helps us to identify **if** and **where** there is a problem. Pay attention that it does **nothing** to **cure** !

Also, the **interpretation** of the results must be taken with great care. Indeed, other cells can highly consume glucose **naturally** (e.g. brain cells, lever cells, etc.).

The "something" that is presented is the strategy is called a **tracer**. Here, the tracer is realized thank to knowledges in **nuclear physics**. 

<div class="alert alert-block alert-info">
The begining of the 20th century is marked by remarkable discoveries in nuclear physics (especially around radioactive decay) that were realized by, to only name a few, Marie Skłodowska-Curie, Henri Becquerel or Ernest Rutherford.</div>

Without entering into the details of this complex physics, a radioactive atom (or molecule) is an **unstable atom** that emits "things" when, due to its unstability, **radioactive disintegrations** occur. 

The **glucose molecule**, whose chemical formula is $C_{6}H_{12}0_{6}$, is modified thanks to a chemical reaction with a **single** radioactive atom so that when the tumor cell "eats" it, the radioactive atom accumulates within the tumor cell itself. Once the radioactive disintegration (also named radioactive decay) occurs, the atom emits two rays of light (called gamma rays). It is by tracking this light that the tumor cell can be located (extremelly preciselly).

The specially **designed** radioactive atom is **fluorine-18** ($^{18}F$). 

Naturally, fluorine has 9 **protons** and 10 **neutrons** ($9+10=19$ elements inside the nucleus - called the nuclei) - it is written $^{19}F$. This way is fluorine **stable** - it is the only way we can find it in nature. It is named a stable **isotope** of fluorine.

<div class="alert alert-block alert-info">
Isotopes are two or more types of atoms that have the same atomic number (number of protons in their nuclei) and position in the periodic table (and hence belong to the same chemical element), and that differ in nucleon numbers (mass numbers) due to different numbers of neutrons in their nuclei. While all isotopes of a given element have almost the same chemical properties, they have different atomic masses and physical properties. Here is an example with the carbon atom :

It is the number of **protons** that makes the element. For example, **oxygen-18** has the same number of **protons** than **oxygen-16**. But suppose for a moment, that a neutron is kicked **out** from the nucleus of oxygen-18 by a impiging proton. By changing the number of protons, we change the element. **Oxygen-18 becomes fluorine-18**. The reaction is written as follows :

$$^{18}O+proton\rightarrow\; ^{18}F+neutron$$

This reaction is represented by the following figure (where electrons are not represented but still present):

This reaction was called **transmutation**.

<div class="alert alert-block alert-info">
Nuclear transmutation was first consciously applied to modern physics by Frederick Soddy when he, along with Ernest Rutherford in 1901, discovered that radioactive thorium was converting itself into radium. At the moment of realization, Soddy later recalled, he shouted out: "Rutherford, this is transmutation!" Rutherford snapped back, "For Christ's sake, Soddy, don't call it transmutation. They'll have our heads off as alchemists." </div>

To realize the transmutation of oxygen-18 into fluorine-18, the proton that is shooted towards the oxygen-18 atom needs a **high energy** (why is this so is not discussed in this session). Approximatelly **18 MeV** (recall that the electron-volt (eV) is an energy unit - see exercise session 7).   

As mentioned above, the radioactive atom is **fluorine-18** (9 protons and 9 neutrons). It is in fact fluorine 19 missing **one neutron**. This way, fluorine is **unstable**. It will eventually emit two light rays that allow us to locate the tumor cell.  

**Fluorine-18** is usually produced by irradiation of $^{18}O$-enriched water ($H_{2}\;^{18}O$) with high-energy (about **18 MeV**) protons prepared in a cyclotron, yielding an **aqueous solution of $^{18}F$ fluorine**. This solution is then used for rapid synthesis of a labeled molecule, often with the fluorine atom replacing a hydroxyl group ($-OH$): a example is presented here with the glucose molecule. Note that the labeled molecules or radiopharmaceuticals have to be synthetized after the radiofluorine is prepared, as the high energy proton radiation would destroy the molecules.

Large amounts of oxygen-18 enriched water are used in **positron emission tomography** centers, for on-site production of $^{18}F$-labeled fludeoxyglucose (FDG) named **fluorodeoxyglucose**.

# Let us accelerate particles : the cyclotron

As we saw during the course, if a **work** is done on a particle, it will change its velocity. The higher the velocity, the higher the energy. As a consequence, to obtain high energy particles, we have to apply a force on them on a certain distance. Pay attention to the fact that certain forces are orthogonal to the displacement and, as a consequence, do not realize any work. 

In this context, let us discuss the **Lorentz force**, expressed as :

$$\vec{F}=q(\vec{v}\times\vec{B})$$

where $q$ is the electric charge of the particle, $\vec{v}$ is the velocity and $\vec{B}$ is the magnetic field. 

For complementary informations, see Reference book (from page 944 to 948).

The particular situation we look at can be figured as follows :  

<img title="a title" alt="Alt text" src="../Figures/Exercises 8/cyclotron.png" width="900">

On the left of the figure, two magnets are facing each other with a spatial gap between them. Within this gap, a peculiar situation occurs : the magnetic field orthogonal to the **orange** plane can be considered as aligned with the $z-$axis and  almost independent of the $x$ and $y$ coordinates. It is represented by the right part of the figure where the magnetic field goes out of your screen (as represented by the dotted circle). The value of the magnetic field is written as $B$, expressed in teslas (T) and depending on the **strength of the magnets**. Additionnal information, the spacial gap where the charges are accelerated is **vacuumed** (otherwise, the particles would collide with a lot of different atoms and molecules).

Thanks to the right-hand rule, we can interpret from the expression of the Lorentz force that a moving particle (with charge $q$ and velocity $\vec{v}$) presents a circular trajectory inside the **orange** plane (as represented on the figure - the charge $q$ is here considered as **positive**). Also, as the force is all the time orthogonal to the displacement, **no work** is realized on the particle. The circular motion is thus **uniform**. 

Given that the trajectory is **know** to be a circle (not demonstrated though), the equation linking the force (mass times acceleration) and the velocity (see Lorentz force) can be written with **scalar** quantities (a vector equation is not mandatory to go further in our computations).

After $n$ turns, the total energy $E$ is :

$$E=n\times \Delta E$$

The value of $\Delta E$ is fixed at the stage of development. Typically, a value of $\Delta E=100$ keV is reasonable. 

6. Using the expressing of the **radius** as a function of the **energy** ($E$), how does the radius changes as a function of the number of turns ?

Mathematically, the number of turns $n$, an **integer**, can be modeled by a **real** number that is an angle expressed (in radians). The relation is the following:

$$n=\frac{\theta}{2\pi}$$

where $\theta$ is an angle. When $\theta=2\pi$, $n=1$, the first turn is complete. When $\theta=4\pi$, $n=2$, etc.
