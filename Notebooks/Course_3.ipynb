# Coordinate system

In course 2, we saw how 1D coordinates system can be useful to study of the motion. Let us go a little further with coordinates systems in 2 and 3 dimensions. 

Here is a painting of Kandinsky (Composition VIII, 1923) that is living on a two-dimensional plane. 

<img title="a title" alt="Alt text" src="../Figures/Cours 3/kandinsky_wassily.jpg" width="600">

All the shapes, lines and curves exist physically independent of any coordinates system. So is true for any physical observations. No observations, nor predictions realized with physical theoretical tools should depend on the way we, observers, choose to write down physics. **The physical laws should be independent of the observers choices.**

Imagine that it would not be the case. That in order to understand the motion of a falling apple, one should measure and write down observations regarding the apple's position with respect to some prescribed location. Or that the physical laws depend on the absolute time (with the origin of time, fixed at the instant of the Big Bang). It would be totally unpractical and absurd. If you want for the laws to be **form invariant**, you must write them with **objects** whose transformation laws are **known** (**prescribed**).

<div class="alert alert-block alert-info">
In theoretical physics, an invariant is an observable of a physical system which remains unchanged under some transformation. Invariance, as a broader term, also applies to the no change of form of physical laws under a transformation. Invariants of a system are deeply tied to the symmetries imposed by its environment. </div>

This course is dedicated to these **objects** and how they change under transformations of interest (namely: **translations, rotations and reflexions**).

Back to our painting, we can see that those shapes, lines and curves may, together reveal some sort of beauty. In human science, things can be said about this painting using **words** (dictionnary, orthography) to create **sentences** (grammar) used in **speeches** (ideas). In physics, the elementary words are **numbers**, sentences are **formulas** and speeches are the **physical laws**. 

To say something about this painting we are going to define a coordinates system (referred to with the **CS** acronym) to **locate** points on this 2D space. To do so, we first have to select an **origin** (**O**). **Distances** between an arbitrary point (**P**) and the origin can now be discussed. 

In addition, we use a CS that is an abstract (sort of) **grid**, considered atop of the painting. Being a hiking lover, let me propose the first CS inspired from **topographic level curves**:

<img title="a title" alt="Alt text" src="../Figures/Cours 3/pointP_a.jpg" width="400">

In this system, the **origin** (white point) is prescribed. The question now is the following: how to communicate the **location** of point **P**? A practical choice could be to use a **finite ordered sequence of numbers**. For points on this painting, we have no choice but to use a **couple** since the painting is two-dimensional. The two coordinates will be referred to in the following way (using the letters $u$ and $v$):

$$(u,v)$$

With this coordinates system, the coordinate $u$ refers to the height value of the **level curve** and $v$ to the angle (in degree) measured between two lines (with the reference "0" prescribed). In such CS, the **sign** of both the "height" and the "angle" **numbers** must be prescribed by an **arbitrary** rule (the height is positive above sea level, the angle grows for anti-clockwise rotations). 

The color code used on this picture is the following: $u$ is constant on the **blue** curves and $v$ is constant on the **red** curves

Where is point **P**? With the given pieces of informations, we can constrain the position of **P** in the region defined by the four points: 

$$(u_1,v_1)=(700,30), \;(u_2,v_2)=(800,30),\; (u_3,v_3)=(800,60), \;(u_4,v_4)=(700,60)$$

- Problem 1 with this CS is that the coordinates do not **uniquely** specify a point. On this picture, it is not the case but in general, the couple ($700,30$), for level curves, could specify arbitrarily many points. 
- Problem 2, how to **construct** level curves in order to locate **precisely** point **P** ? 

Here is another "CS" attempt: 

<img title="a title" alt="Alt text" src="../Figures/Cours 3/pointP_b.jpg" width="400">

What a chance ! We can now be precise. **P** is located in $(u,v)=(2,1)$. Note that with this CS, the origin is associated with the coordinates $(u,v)=(0,0)$. With the previous system, it was not unique. Indeed, $(u,v)=(0,12.4)$ or $(u,v)=(0,-37.3)$ refer equivalently to the origin. 

So, why don't we use this CS as a **textbook reference** ? (try to guess first)

Because there is no "natural" **directions** given with this CS. Let us try to construct such **directions** at point **P**. 

<img title="a title" alt="Alt text" src="../Figures/Cours 3/local_axes.jpg" width="500">

The **geometrical** construction needed to built these directions **at point P** is the following:

Construction on Figure (a):
- Construct the arrow linking the origin $(u,v)=(0,0)$ and point P $(u,v)=(u^{*},v^{*})$.
- Construct the arrow linking the origin (O) and point Q. Point Q $(u,v)=(u^{*}+\Delta u,v^{*})$ is located somewhere further away where **only $u$** changes.
- Construct the arrow linking point P and point Q.
- Divide this arrow by the variation $\Delta u$ (not shown).

Construction on Figure (b):
- Let $\Delta u$ tend towards 0.
- Construct the arrow linking point P and point Q.
- Divide this arrow by the variation $\Delta u$ (not shown).

This **geometrical** process is **convergent** and can be reproduced in a process where **only $v$** changes from $P$. This gives us two **local directions**. The problem is that, as you can see on Fig. (c), these directions are specific to point $P$ and vary from point to points. 

To solve this problem, here is the CS that many textbook present:

<img title="a title" alt="Alt text" src="../Figures/Cours 3/pointP_c.jpg" width="400">

The **great advantage** is that the geometrical process presented above is again convergent and produce the **same** directions **anywhere** (no matter if you consider the green, yellow, purple, white, etc. points). In addition, due to the regularity of such grid, the produced arrows should be the **same, anywhere**. Due to this regularity, it is now possible to be precise on the coordinates of P **even if** the coordinates do not take integer values. P is located at $(u,v)=(1.5,4)$. This specific CS is called a **Cartesian coordinate system**. The coordinates have "reserved" notations that are $(x,y)$ instead of $(u,v)$. 

<div class="alert alert-block alert-info">
In a plane is a Cartesian coordinate system specifies each point uniquely by a pair of numerical coordinates, which are the signed distances to the point from two fixed perpendicular oriented lines, measured in the same unit of length. Each reference coordinate line is called a coordinate axis or just axis (plural axes) of the system, and the point where they meet is its origin, at ordered pair (0, 0). </div>

The arrows that are **geometrically** constructed are called **unit vectors** because they are associated with the length (called **norm**) **1**. In this course, they are noted: $\textbf{e}_{x}$ (direction along which $x$ varies) and $\textbf{e}_{y}$ (direction along which $y$ varies). 

Their formal definition is given as follow and implies **partial derivatives** (considered as prerequisit - see INGE1133). The (abstract) object associated with **P** is noted $\vec{P}$ and the geometrical procedure depicted above leads to the following definitions:

$$\textbf{e}_{x}=\frac{\partial\vec{P}}{\partial x}$$

$$\textbf{e}_{y}=\frac{\partial\vec{P}}{\partial y}$$
 
The object associated to **P** is written:

$$\vec{P}=1.5\textbf{e}_{x}+4\textbf{e}_{y}$$

**Geometrically**, we can interpret the **1.5** as the answer to the following question: How much of $\vec{P}$ points along $\textbf{e}_{x}$? 

Thanks to **Pythagorean theorem** we can define the **length** (called the **norm**) of $\vec{P}$. Since $\vec{P}$ is the hypotenuse of a right triangle, we have:

$$|\vec{P}|^{2}=(1.5)^{2}+(4)^{2}$$

This computation allows us to answer geometrically to the following question: How much of $\vec{P}$ points along an **arbitrary unit vector** pointing along $\vec{Q}$? Such unit vector is noted $\textbf{e}_{Q}$ and is given by:

$$\textbf{e}_{Q}=\frac{\vec{Q}}{|\vec{Q}|}$$

The coordinates of the point **Q** are $(x,y)=(4,3)$ so,

$$\vec{Q}=4\textbf{e}_{x}+3\textbf{e}_{y}$$

See the following figure:

<img title="a title" alt="Alt text" src="../Figures/Cours 3/scalar_product.jpg" width="400">

The green line has the following lenght $p_{q}$ (a simple product between two numbers):

$$p_{q}=|\vec{P}|\times cos(\theta)$$

where $\theta$ is the angle (in radians) between $\vec{P}$ and $\vec{Q}$. This quantity is **extremelly** important in physics so let us **give it a name**: the **scalar projection**.

<div class="alert alert-block alert-info">
The scalar projection (or scalar component) of a Euclidean vector $\vec{P}$ in the direction of a Euclidean vector $\vec{Q}$ is given by the quantity $|\vec{P}|\times cos(\theta)$.</div>

This quantity can rewritten in term of an object that we  will **define** later in these notes, namely, the **scalar product** (also called the **dot product**). We have:

$$p_{q}=\vec{P}\cdot \textbf{e}_{Q}$$

Using the definition of $\textbf{e}_{Q}$, we have (**important!**):

$$\vec{P}\cdot\vec{Q}=|\vec{P}||\vec{Q}|cos(\theta)$$

This definition guides us towards a "natural" **algebra** (**algebraic structure**). We have:

$$\textbf{e}_{x}\cdot\textbf{e}_{x}=1\times 1\times cos(0)=1$$

$$\textbf{e}_{y}\cdot\textbf{e}_{y}=1\times 1\times cos(0)=1$$

and

$$\textbf{e}_{x}\cdot\textbf{e}_{y}=\textbf{e}_{y}\cdot\textbf{e}_{x}=1\times 1\times cos(\frac{\pi}{2})=0$$

We can now write:

$$\vec{P}\cdot\vec{Q}=(1.5\textbf{e}_{x}+4\textbf{e}_{y})\cdot(4\textbf{e}_{x}+3\textbf{e}_{y})$$

It is left as an exercice to show that the **scalar product** satisfy a **distributive law** so we have:

$$\vec{P}\cdot\vec{Q}=(1.5\times4)(\textbf{e}_{x}\cdot\textbf{e}_{x})+(1.5\times 3)(\textbf{e}_{x}\cdot\textbf{e}_{y})+(4\times 4)(\textbf{e}_{y}\cdot\textbf{e}_{x})+(4\times 3)(\textbf{e}_{y}\cdot\textbf{e}_{y})=(1.5\times4)+(4\times 3)=18$$

In order to be completely general, let us write the coordinates of **arbitrary** points $(x_{p},y_{p})$ (for point **P**) and $(x_{q},y_{q})$ (for point **Q**). We have that the scalar product (in two dimensions) is given by:

$$\vec{P}\cdot\vec{Q}=x_{p}x_{q}+y_{p}y_{q}$$

<div class="alert alert-block alert-danger">
At this point of the notes, we did not defined what a <b>scalar</b> is, nor the notion of <b>vector</b>. Nonetheless, as most of you noticed, such objects already appeared in these notes. e.g. $\vec{P}$ is a vector and $\vec{P}\cdot\vec{Q}$ is a scalar.
</div>

One can now ask if these definitions are useful to manipulate describe **reality**. Yes, yes indeed (and you will see how in the rest of the course).

Let us first try to use these tools to describe the point **P** in a CS that is rotated by an angle $\theta$ about the origin (starting from the CS presented above).

<img title="a title" alt="Alt text" src="../Figures/Cours 3/rotation.jpg" width="600">

The "old" coordinates are $(x,y)$. The "new" ones are $(x',y')$. How are they related to each other? Let's find out.

First, how are the "new" **unit vectors** related to the "old" ones. We know that:

$$\textbf{e}_{x'}=\frac{\partial\vec{P}}{\partial x'}$$

$$\textbf{e}_{y'}=\frac{\partial\vec{P}}{\partial y'}$$

Using the **chain rule** (see INGE1133) we have:

$$\textbf{e}_{x'}=\frac{\partial\vec{P}}{\partial x'}=\frac{\partial\vec{P}}{\partial x}\frac{dx}{dx'}+\frac{\partial\vec{P}}{\partial y'}\frac{dy}{dy'}=\textbf{e}_{x}\frac{dx}{dx'}+\textbf{e}_{y}\frac{dy}{dy'}$$

What is $\frac{dx}{dx'}$ ? Using the following triangles (that you can see in the previous figure), 

<img title="a title" alt="Alt text" src="../Figures/Cours 3/trigo.jpg" width="300">

it is easy to see that:

$$\frac{dx}{dx'}=cos(\theta)$$

With similar arguments, it is clear that:

$$\frac{dy}{dx'}=sin(\theta)$$

and so,

$$\textbf{e}_{x'}=cos(\theta)\;\textbf{e}_{x}+sin(\theta)\;\textbf{e}_{y}$$

It is left as an exercice to show that:

$$\textbf{e}_{y'}=-sin(\theta)\;\textbf{e}_{x}+cos(\theta)\;\textbf{e}_{y}$$

By definition, the quantity we are looking for ($x'$) answers the following question: how much of $\vec{P}$ points along $\textbf{e}_{x'}$ ? We can use the **scalar product** to find the answer:

$$x'=\vec{P}\cdot\textbf{e}_{x'}=(x\;\textbf{e}_{x}+y\;\textbf{e}_{y})\cdot(cos(\theta)\;\textbf{e}_{x}+sin(\theta)\;\textbf{e}_{y})$$

that is

<div class="alert alert-block alert-info">
$$x'=x\;cos(\theta)+y\;sin(\theta)$$

and similarly,

$$y'=-x\;sin(\theta)+y\;cos(\theta)$$ </div>

The key indicator that these definitions could be useful in physics is that the **norm** of $\vec{P}$ **does not change under rotation**:

$$|\vec{P}|^{2}=x^{2}+y^{2}=x'^{2}+y'^{2}$$

This result is really important ! Suppose that $\vec{P}$ the abstract representation of a rod, this relation shows that its **length** would be the **same** in **any** rotated frame (which is really good news). 

# Translation, rotation and reflection

The definition of a vector is central to the course. Many textbook define this object as a quantity that **has magnitude and direction**. This definition is unsatisfactory because, it is a direction with respect to what ? Who **gives** you this direction ? In this course we use another definition.

### Vector definition

- A vector is an object whose components vary according to the same rule as the coordinates themselves. 
- A vector is **invariant** (its components do not change) under **translation** (by construction).
- Its **norm** is invariant under **rotation** and reflection.

### Operations with vectors

- Vectors can be added (or subtracted) component-wise. If $\vec{a}$ has components in a carthesian CS $(a_{x},a_{y})$ and $\vec{b}$ has components $(b_{x},b_{y})$ then $\vec{a}+\vec{b}$ has components $(a_{x}+b_{x},a_{y}+b_{y})$
- There exists two "products":
  - the scalar product (or dot product), producing a scalar, $\vec{a}\cdot\vec{b}$ defined as:         $$\vec{a}\cdot\vec{b}=a_{x}b_{x}+a_{y}b_{y}$$
  - the vector product (or cross product), producing a vector colinear with the vector  ($\vec{n}$) normal to the plane defined by $\vec{a}, \vec{b}$: $$\vec{a}\times\vec{b}=(a_{x}b_{y}-a_{y}b_{x})\vec{n}$$
 
### Scalar definition

A scalar is a quantity that is **invariant** (does not change) under **translation, rotation or reflection**.
