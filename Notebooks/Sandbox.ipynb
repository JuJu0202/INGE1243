In a typical Wimbledon tennis tournament, 42.000 tennis balls are used. The offical diameter of a tennis ball must be around 6.7 cm. If we lay all these balls on the ground, does the covered area fits inside the playing area of a single tennis court (a court is a rectangle of 23.77 m of length and 10.97 m of width) ? (expected answer : 'yes' or 'no') (a) If the balls are packed on a square lattice. (b) If the balls are packed on an hexagonal lattice.

```python
import numpy as np

# Number of balls per tournament

N = 42000

# Tennis ball radius

R = (6.7e-2)/2 # [meter]

# In a square lattice, there is one ball per unit cell
f_sq = 1 # [ball/unit]
# In an hexagonal lattice, there are two balls per unit cell
f_hex = 3 # [balls/unit]
# The area of the square
a_sq = (2*R)**2  # [meter square/unit]
# The area of the hexagon
a_hex = (3*np.sqrt(3)/2)*(2*R)**2  # [meter square/unit]
# The density is given by
d_sq = f_sq/a_sq # [(ball/unit)/(meter square/unit) = (ball/meter square)]
# The density is given by
d_hex = f_hex/a_hex # [(ball/unit)/(meter square/unit) = (ball/meter square)]


# This gives us a total area (A) of :
A_sq = N/d_sq # [meter square]
# This gives us a total area (A) of :
A_hex = N/d_hex # [meter square]

# The playing area of tennis court is given by:

S = 10.97*23.77 #[meter square]

print("The area covered by a square packing is %.2f m², which is bigger than the %.2f m² for an hexagonal packing" % (A_sq, A_hex))
print("Both fits in the playing area of %.2f m²" % (S))
```
