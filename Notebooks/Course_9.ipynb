Being so vast, it comes with a lot of scientific terms but in the begining, let's start with only two : **electric charge**, **electric fields**.

### Electric charges

It is known for a fact that there are **two kinds** of electric charges. 

1. **Like** charges **repel** each other
2. **Opposite** charges **attract**

This very fact is already puzzling. Remember that, historically, the gravitational force was the first one to be understood. We saw that particles with masses **always** attract and that the magnitude of this force was :

$$F=G\frac{m_{1}m_{2}}{R^{2}}$$

where $G$ is the gravitational constant ($6.6\times 10^{-11}$ $[Nm^{2}kg^{-2}]$), $m_1$ the mass of the first particle, $m_2$ the mass of the second particle and $R$, the distance between the two. 

Recall that, by Newton's third principle, the first particle applies the **same** force (whose magnitude is $F$) on "particle 2" than "particle 2" on "particle 1". We can thus construct the following figure :

<img title="a title" alt="Alt text" src="../Figures/Cours 9/gravity.png" width="400">

It was also noted that, charged object can repel and attract each other, which causes **motion**, meaning an **acceleration** and thus, a **force**.

The different configurations are presented on the following figure : 

<img title="a title" alt="Alt text" src="../Figures/Cours 9/charge.png" width="600">

To mathematically take this fact into account, the charge of a particle is specified with its sign. A **plus** sign for a **positive** charge and a **minus** sign for a **negative** sign. The four presented situations are then taken into account by a product between the charges :

1. "+" times "+" equals "+"
2. "+" times "-" equals "-"
3. "-" times "+" equals "-"
4. "-" times "-" equals "+"

The mathematical formula for the **electrostatic force** is called **Coulomb's law** (named after the French physicist Charles Coulomb (1736-1806)) and is given by:

$$F=k\frac{|q_{1}q_{2}|}{R^{2}}$$

The new physical dimension encountered here, the electrical **charge** is expressed in **Coulomb** in the SI system and is written $[C]$ (like we write [kg] for a mass). In SI units, the constant $k$, present in this equation, is measured to be:

$$k=9\times 10^{9}\;[\frac{N m^{2}}{C^{2}}]$$

All trough the course, it will be clear that there is a deep analogy between the gravitational force and the electrostatic force. Nevertheless, there is a striking difference between these two forces: their difference in magnitude !

Suppose you have two protons one meter appart from each other. The proton's mass is approximatelly $1.6\times 10^{-27}$ kg and its charge is (the opposite of the charge of the electron) $1.6\times 10^{-19}$ coulomb. The magnitude of the gravitational force ($F_{g}$) that would bring them closer together is (in order of magnitude):

$$F_{g}\approx 10^{-11}\times \frac{10^{-27}\times 10^{-27}}{1^{2}}\approx 10^{-65}\; Newtons$$

On the other hand, the electrostatic force $F_{e}$ causing them to repel each other is given by :

$$F_{e}\approx 10^{10}\times \frac{10^{-19}\times 10^{-19}}{1^{2}}\approx 10^{-28}\; Newtons$$

$F_{e}$ is a small force... but $F_{g}$ is immensely smaller ! As we learned from previous courses, the net force acting on a body is the sum of the forces but in this case, the gravitational interaction can be completely neglected ! This fundamental difference is still intriguing generations of scientists today. 

Looking at numbers, there is another interesting fact to note, the elementary particles carrying an electric charge are the proton and the electron. Electricity is about the way charges move and arrange spacially in peculiar configurations. The movement of these objects is characterised by their **acceleration**. Thanks to Newton's second law, we know that the acceleration is the force divided by the mass. The rest mass of an electron is $9.1\times 10^{-31}$ kg where the mass of a proton, as already stated, is $1.6\times 10^{-27}$ kg. The electron is  approximately **1836** times lighter than the proton. For the same force acting on both, the distance covered by the electron is then all the time much bigger. This is why the effects we deal with in our daily life are caused by **electrons** (we talk about **elec**tricity and **elec**tronics and not **proton**icity or **proton**ics).
