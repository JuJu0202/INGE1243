# Magnetism (Part 1)
 
In the previous course, we discussed how the presence of electric **charge** was linked to the presence of a electric **field** in the space all around it. We found that a charged particle $q$ creates an electric field $\vec{E}$ in the space all around it.

In this course we will see that there exists another **field**, called the **magnetic field**, whose presence is also causing a force on a charged particle. 

The discovery of the magnetic field is associated with the existence of **magnets**. The earliest known surviving descriptions of magnets and their properties are from Anatolia, India, and China around 2500 years ago.

It is known for a fact that **any** magnets are constituted from **two kinds** of poles. 

1. **Like** poles **repel** each other
2. **Opposite** poles **attract**

It was also discovered that **permanent** magnets could **magnetize** objects with magnetic properties. For example, **iron** can be magnetized by magnets and becomes a magnet itself. For this reason, **iron filings** is used for educational and research purposes to study the spacial distribution of the magnetic field. Iron filings are very small pieces of iron that look like a light powder. 

The space around the magnet contains a **magnetic field** written $\vec{B}$. The shape of the field is revealed by the filings, which **align** with the magnetic field lines that spread out from one pole and return to the other. By **convention**, these lines are **oriented** from the North to the South pole. The iron powder is sprinkled on a plate where a magnet is deposited. Conventionaly, the opposite poles are called the **north** (noted with the letter N) and the **south** (noted with the letter S) pole. Without telling you why, just know that even if you break a magnet in two parts, you cannot (up to our present knowledge) create magnetic monopoles (a single North or a single South). The two, freshly created objects, present 2 magnetic poles. You now have 2 smaller magnets. </div>

The fact that a magnet always presents a pair of poles makes it a facinating study case for scientists of all kind. A first remarkable point is that :

Whereas **electric charges** placed in an electric field **accelerate**, **magnets** placed in a magnetic field only **align**.

Let us take a minut to see how such reality is **modeled** for magnets. When a magnet is placed in a magnetic field, it wants to rotate **about** its center of mass. Here is a figure of a **compass** present in a magnetic field (pictured by the parallel oriented dashed lines on the figure):

<div class="alert alert-block alert-info">
By the 12th to 13th centuries AD, magnetic compasses were used in navigation in China, Europe, the Arabian Peninsula and elsewhere. Note that, if the compass needle points towards the geographic north pole of the Earth, it means that, from a magnetic point of view, the geographic north pole... is a south magnetic pole ! In addition, know that the vector connecting the geographic south pole to the north pole is not colinear with the vector connectic the magnetic south pole to the north one. In addition, Earth experiences a slow (really slow), yet measurable, pole inversion. The orientation of the Earth magnetic field is not permanently fixed.</div>

This observation tells us that the magnet experiences a net **torque**. It is in fact this torque that we have to model. To do so, let us consider that a **vector** (called the magnetic moment - written $\vec{\mu}$) is associated with the magnet. The larger the magnitude of $\mu$, the stronger the magnet. This vector is aligned with the virtual oriented line that connects the south to the north pole. 

<div class="alert alert-block alert-danger">
Pay attention to the fact that $\vec{mu}$ is not a force, nor is $\vec{B}$. Both this quantities have their own dimension and units. </div>

The torque is given by the following equation :

$$\vec{\tau}=\vec{\mu} \times \vec{B}=|\vec{\mu}||\vec{B}|sin(\theta)\textbf{e}_{z}$$

where $\theta$ is defined on the following figure as the angle between the magnetic moment $\vec{\mu}$ and the magnetic field $\vec{B}$. The $"\times"$ refers to the **cross product** (right-hand rule). The sign of $\tau$ is important to signify the rotation orientation (a positive $\tau$ (pointing outside of your screen) translates into an anti-clockwise rotation where a negative one translates into a clockwise rotation - try to play with this law.)

Now that we treated the "vector" part, let us have a look at energy consideration. We know that the infinitesimal **work** ($dW$) associated to a **torque** ($\tau$) is given by :

$$dW=|\vec{\tau}| d\theta$$

It is mandatory to consider infinitesimal works because after each $d\theta$ movement of the magnet, $\tau$ changes ! The total work $W$ is the **sum** of all these infinitesimal works (a.k.a. an **integrale**).

$$W=\int_{\theta_1}^{\theta_2} dW=\int_{\theta_1}^{\theta_2}|\vec{\tau}| d\theta=\int_{\theta_1}^{\theta_2}\mu B sin(\theta) d\theta=-\mu B\left[cos(\theta)\right]_{\theta_1}^{\theta_2}$$

The work required to flip a magnet from **state 1** (unstable equilibrium - $\theta_{1}=\pi$) to **state 2** (stable equilibrium - $\theta_{2}=0$) is given by :

$$W=-\mu B (cos(0)-cos(\pi))=-2\mu B$$

In term of potential energy **P.E.**, we have :

$$P.E.=W=-\mu B cos(\theta)= -\vec{\mu}\cdot\vec{B}$$

where the $"\cdot"$ is a **dot product** (...how elegant). 

### The magnetic field due to a magnet : a brief comment

As mentioned, the space around the magnet contains a **magnetic field** written $\vec{B}$. This magnetic field depends on the distance $d$ between the point where the field is evaluated and the center of mass of the magnet. It also depends on the magnetic moment $\mu$. Without going into the total expression of the field anywhere in the space around the magnet, let us look at the field colinear with the magnetic moment $\mu$ expressed as $B_{axis}$

<img title="a title" alt="Alt text" src="../Figures/Cours 10/baxis.png" width="700">

The magnetic field, $B_{axis}$ (measured in **tesla**), of an ideal dipole measured along its axis is :

$${B_{axis}} = \frac{{{\mu _0}}}{{4\pi }}{\text{ }} \frac{{2\mu }}{{{d^3}}}$$

where µ0 is the permeability constant ($4\pi 10^{-7} [T m/A]$), d is the distance from the center of the dipole in meters and µ is the magnetic moment. The magnetic moment, µ, measures the strength of a magnet, much like electrical charge measures the strength of an electric field source.

Now that you have an idea of **how** the magnetic **field** can vary on a given **field line**, let us look (with no formulas) at the repartition of the magnetic field lines in **magneto-static** configurations (all source of magnetic field are static - not moving).
