### Work

On the following figure you can see two workers that need to bring a bag of cement (10 kg) to the first floor of a building that is in construction.

<img title="a title" alt="Alt text" src="../Figures/Cours 7/work.jpg" width="800">

The worker on the left of the figure is named Tom and the one on the right, Jerry. Tom took **6 min** to bring this bad at the level of the first floor whereas Jerry took **15 min**. 

They both **worked** hard to bring these two bags on the first floor. Let us consider the following definition for **work**.

Work is the effort exerted **on** something that will **change its energy**. The **total energy** of an object will be written $E$. 

Two things enter the picture whenever work is done: 
1. the application of a force 
2. the movement of something by the action of that force. 

For the simplest case, where the force is constant and the motion is in a straight line in the direction of the force, we define the work done on an object by an applied force as the product of the force and the distance through which the object is moved. In shorter form:

$$Work\;=\;Force\;\times\;distance$$

or

$$W=F\times d$$

Work has the physical dimension $[W]$=$[F][d]=[ML^{2}T^{-2}]$ (an energy). The unit of energy is called the Joule $[J]=[kg\frac{m^{2}}{s^{2}}]$

Let us use this definition for the two situations:

### Tom's bag

Here is the situation:

<img title="a title" alt="Alt text" src="../Figures/Cours 7/tombag.jpg" width="400">

**Before** any work, the tension in the rope is zero and the forces acting on the bag are:

- Gravity = $\vec{F}=-mg\textbf{e}_{y}$
- Floor's reaction (preventing the bag to dig into the Earth): $\vec{n}=+mg\textbf{e}_{y}$

To lift the bag, Tom needs to exert a force **greater than** $+mg\textbf{e}_{y}$ (otherwise nothing moves). Once in motion, the bag can maintain a steady velocity even if the force **equals** $+mg\textbf{e}_{y}$. Let's consider that Tom exerts a force of magnitude $mg$ on the distance $d$ that is the height of the first floor $h$. According to the definition, the work realized **by** Tom to lift the bag ($h=3.5\;m$) is : 

$$W_{Tom}=mgh\approx 343\;[J]$$

Tom physicaly **changed** the energy of the bag by the amount $mgh$ thanks to his work. We considered that Tom used a frictionless pulley to transmit his force. 

Let's now have a look to Jerry's bag.

### Jerry's bag

<img title="a title" alt="Alt text" src="../Figures/Cours 7/Jerrybag.jpg" width="400">

To move his bag (we neglect the bag's friction on the floor), Jerry needs to exert a force $-mgsin(\theta)\textbf{e}_{x'}$ along a distance $d$ given by the length of the ramp (otherwise the bag slides back towards the ground).

$$d=\frac{h}{sin(\theta)}$$

We thus have :

$$W_{Jerry}=mgsin(\theta)\times\frac{h}{sin(\theta)}=mgh\approx 343\;[J]$$

<div class="alert alert-block alert-info">

It seems that (not demonstrated here, yet true !), **no matter the path** choosen to lift the bag to the first floor, the **energy change** of the bag is **always the same** : $mgh$. Also, the computation for the energy change (work) in simple situations is quite a simple formula. If you double the mass (20 kg), you double the work. If you double the height, you double the work. Easy. 

This concept is cool but now comes the crutial question : Who do you think delivered the largest amount of energy to lift the bag to the first floor ? Tom or Jerry ? (Stop reading and think about the answer).

Considering the **work** realized on the bag, we saw above that it was exactly the same so your first answer could have been : Tom delivered an energy $E_{Tom}$ that is equal to $E_{Jerry}=W_{Jerry}=W_{Tom}$. 

In fact, this is **not** true because : 
 
1. Jerry also has to **lift** himself to the first floor (which is not an energy free process). Assuming Jerry has a mass $M$ of 80 kg, the delivered energy is $(m+M)gh\approx 3087\;[J]$
2. We neglected **friction** (discussed later).

The energy that the box gained between the ground and the first floor is called the **potential energy** (PE). Here, the bag **stored energy** that can be **potentially** retrieved. Such energy is stored by vertue of its position (height) : it is gravitational potential energy. 

But the potential energy is not always due to gravity. Can you think of examples ?

# Kinetic energy

There are other forms of energy - the energy you can set in the **motion** of objects, namely, **kinetic energy** (KE). The kinetic energy of an object depends on the mass ($m$) of the object as well as its speed $v$. Using **dimensional analysis**, we know that the only way to construct an object with the physical dimension of an energy is :

$$[E]=[m]^{a}[v]^{b}$$

with $(a,b)=(1,2)$.

The exact formula is :

$$Kinetic\;energy\;=\frac{1}{2}\;mass\times speed^{2}$$

This energy is the amount required to bring the object to its velocity $v$ from rest. Let us apply this concept to the situation of a tennis ball that is dropped from rest from a height of 1 meter. The figure is the following : 

<img title="a title" alt="Alt text" src="../Figures/Cours 7/tennis.jpg" width="300">

To lift the ball to an height $h$ (1 m), we know that the potential energy (PE) of the ball is $mgh$. Once it reaches the floor, since $h=0$, the ball has no PE left ($PE=0$). The question is then, what is the velocity ($v$) of the ball (to compute its Kinetic Energy) ? 

We know that the time interval $\Delta t$ to cover a distance $h$ with a constant acceleration $g$ is given by :

$$h=\frac{g\Delta t^{2}}{2}\leftrightarrow \Delta t=\sqrt{\frac{2h}{g}}$$

The velocity after such time interval is :

$$v=g\Delta t=\sqrt{2gh}$$

The KE is then:

$$KE=\frac{1}{2}mv^{2}=\frac{1}{2}m 2 g h=mgh$$

This fact is at the root of the **work energy theorem**. A **change** in KE, from 0 to $\frac{1}{2}mv^{2}$, is **work** ! To change the KE of an object, a **net** force is required on a distance $d$. Let us consolidate this concept (Work equals change in KE: $W = \Delta KE$) with an example.

Suppose you ride your bike at a velocity $v$ along a straight road and you notice a **STOP signal**. You directly and firmly apply your brakes (causing the wheel to stop rotating). The friction between the tires and the road is given as half the weight of you and the bicycle. 

The question is then : How far will you skid ? 

First, when you encounter such problem, make a sketch of the situation : 

<img title="a title" alt="Alt text" src="../Figures/Cours 7/bike.jpg" width="450">

The force acting on the bike are : gravity and floor's reaction (balancing)(in red on the figure) + friction. Since there are two tires, the reaction force is half the weight (at each tire) and we know that the friction force is considered to be constant and half the weight (so a quarter at each contact point between the tire and the floor). 

The answer to "how far ?" is a **distance** $d$. We also know that the change in kinematic energy must be **all** the KE (in order to reach a velocity of zero). We have:

$$\Delta KE=\frac{1}{2}mv^{2}- 0$$

Which must be equal to a work. Which one ? The work corresponding to the friction (braking) force $F_{b}$ (said to be $\frac{1}{2}mg$) on the distance $d$ (the unknown). We can write : 

$$F_{b}\times d=\Delta KE=\frac{1}{2}mv^{2}$$

This gives you:

$$d=\frac{v^{2}}{g}$$

The central assumption here is that the friction force $F_{b}$ is constant **no matter** the instanteneous velocity. Actually, this assumption is quite valid. The friction force is almost the same if you were riding at 25 km/h or 50 km/h. But as you can see, if you double the speed, the braking distance is multiplied by a factor 4 ($2^2$) ! 

So, as accident investigators are well aware, an automobile going 100 km/h has four times the kinetic energy that it would have
at 50 km/h, and the car skids four times as far with its wheels locked as it would going 50 km/h. 

Now that we defined the concept of work, it is interesting to compare the work corresponding to the friction force with the work required to lift the bag (presented earlier). 

Indeed, suppose a work is realized to lift the mass presented on the following figure (in red). 

<img title="a title" alt="Alt text" src="../Figures/Cours 7/potential.jpg" width="450">

As discussed the work equals $mgh$ and is PE. If the rope is snapped at some instant, the mass will fall and this energy converts progressively into KE. This KE can then be **used** to **plant a nail** for example. This energy can be retrieved to realize **another usefull work**. Potential energy is all around us to help us in our daily life to realize... work. 

Let us compare it with the work corresponding to the friction force on the braking distance: $F_{b}\times d$. Can we retrieve this energy to realize useful work ? The answer is **no**, unfortunately. This work converts into **thermal energy** and is simply **dissipated** : No further useful work is possible; the energy is now 'lost to the environment'. Look at this infrared thermal imaging of the tire taken just after braking. 

<img title="a title" alt="Alt text" src="../Figures/Cours 7/braking.jpg" width="450">

This thermal energy flows as **heat** within the tire and the floor, ultimately raising the temperature of both of these objects. A whole branch of physics, thermodynamics, deals with how heat is transferred between different systems and how **work**, if possible, is done in the process. Thermal engine being the clearest example of the interplay between these two concepts : work and heat.

### Conservation of energy

The important thing in all these example is not the philosophical debate on the **nature** of energy but rather how it evolves in physical objects. 

The study of various forms of energy and their transformations from one
form into another has led to one of the greatest generalizations in physics—the
law of conservation of energy:

**Energy cannot be created or destroyed; it can be transferred from one
object to another or transformed from one form into another, but the
total amount of energy never changes.**

Let us consolidate this concept with an example that we already encountered : the projectile motion. 

### The projectile motion : energy conservation

Let us consider the following figure :

<img title="a title" alt="Alt text" src="../Figures/Cours 7/projectile_motion.jpg" width="600">

Let us consider that the object of mass $m$ has no p\frac{1}{2}mv^{2}=otential energy in the begining of its motion ($t=0$). Doing so, we **arbitrarily determine** the PE **reference**.

The total energy of the particle (purely KE) is :

$$E=\frac{1}{2}mv^{2}=\frac{1}{2}mv_{0x}^{2}+\frac{1}{2}mv_{0y}^{2}$$

We know that a **change** in KE is a **work**. Whose **work** ? The one associated with the gravitational force on the trajectory. Let us consider another point of interest : The highest point (middle red point on the figure). As mentioned earlier in the course, **no matter the path** choosen to rise the mass, its **energy change** is **always the same** : $mgh$ (with $h$ the maximal height of the particle - unknown).

To figure out what $h$ is, let us **note** the following point :

$$\Delta KE=Work$$

What changed considering the KE ? What is $\Delta KE$ ? At the highest point, the vertical component of the velocity is **zero**. The **new** KE is given by:

$$\frac{1}{2}mv_{x}^{2}+\frac{1}{2}mv_{y}^{2}=\frac{1}{2}mv_{0x}^{2}$$

Where did $\frac{1}{2}mv_{0y}^{2}$ go ? It could not simply disappear... it can only be transferred from one from one form into another. From Kinetic to Potential. So we have : 

$$mgh=\frac{1}{2}mv_{0y}^{2}$$

This equation **defines** $h$ : 

$$h=\frac{v_{0y}^{2}}{2g}$$

We can now wonder what is the time interval $\Delta t$ required to reach such height $h$ ?

Easy ! We know that :

$$a_{y}=-g$$

$$v_{y}=v_{0y}-g\Delta t$$

and eventually :

$$y=v_{0y}\Delta t-\frac{g\Delta t^{2}}{2}$$

At the highest point, $y=h$ so :

$$\frac{v_{0y}^{2}}{2g}=v_{0y}\Delta t-\frac{g\Delta t^{2}}{2}$$

Multiplying both side of the equation by $2g$ and putting all the terms on the same side of the equation, we have the equation **defining** $\Delta t$:

$$g^{2}\Delta t^{2}-2gv_{0y}\Delta t+v_{0y}^{2}=0$$

Which can be rewritten as :

$$(g\Delta t-v_{0y})^2=0$$

Giving,

$$\Delta t = \frac{v_{0y}}{g}$$

Let us go further : what is the total range of the shoot ? (the distance $d$ when the particle re-reached the height $y=0$, where the particle has zero potential energy). 

To reach this distance, the particle takes **twice the time** it took to reach the highest point (by symmetry - $2\Delta t$). We also know that horizontally, the horizontal component of the velocity does not change (momentum conservation - there is no force with a component along "x" acting on the particle). The distance is simply :

$$d=v_{0x}\times 2\Delta t=\frac{2v_{0x}v_{0y}}{g}$$

We also know that :

$$v_{0x}=vcos(\theta_{0})$$

and 

$$v_{0y}=vsin(\theta_{0})$$

Eventually : 

$$d=\frac{2v_{0x}v_{0y}}{g}=\frac{2v^{2}sin(\theta_{0})cos(\theta_{0})}{g}=\frac{v^{2}sin(2\theta_{0})}{g}$$

Which is totally consistant with what we got with both **dimensional analysis** and **Exercise 6 (exercise session 3)**.
