Considering the motion of objects, one can distinguish two categories: on the one hand, arbitrarily complex, irregular motions and on the other, motions where coordinates and dynamical variables come back to their starting values after same time. 

In course 4, we encountered a first type of periodic motion, namely, the **uniform circular motion**. In course 8, we described another periodic motion : the **simple pendulum**.

In this exercise session, let us focus on another object where a **periodic** motion is at play : the **spring**. Here is a representation of such spring.

<img title="a title" alt="Alt text" src="../Figures/Exercises 6/spring.png" width="400">

It is known, as an experimental fact that, if you pull on the spring from its "rest length" $L$, a distance $x$ (from $x=0$), there is a restoring force **proportional** to the displacement. For exemple, if you triple $x$, you triple the force etc. 

When $x$ is positive, the force points in the negative direction and when $x$ is negative, the force is in the positive direction. This can be **modeled** by the following equation :

$$F(t)=-kx(t)$$

where $k$ is called the **spring constant**. This law is refered to as **Hooke's law**.
