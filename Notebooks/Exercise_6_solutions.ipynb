4. There are no single answer to this question. It can be either the blue or the green curve.
    1.     2. If you assumed that friction is **huge**, the blue curve is the one. Indeed, the mass can never cross its equilibrium position with a non-zero velocity. The restoring force brings the mass back to its equilibrium position with no oscillation. The spring is said to be **critically damped**.
    3. In physics related situations, the damping is often not that big and for a reasonable amount of time, the oscillatory motion can be considered as a good **model**.
